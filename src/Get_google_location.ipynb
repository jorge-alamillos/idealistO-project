#### En este Jupyter generamos la posición geográfica (latitud y longitud) para cada inmueble del dataset. Para ello recurrimos a la API de GeoCode de Google.

```python
import time
import pandas as pd
from dotenv import load_dotenv
import requests
load_dotenv()
import os
```

Importamos el dataset de inmuebles tal y como lo descargamos

```python
rstate = pd.read_csv("houses_Madrid.csv")
```

Eliminamos aquellas filas del dataset que no contienen la dirección, puesto que sin ella no podemos obtener las coorednadas exactas, siendo inútiles para nuestro trabajo.

```python
rstate.dropna(subset=["raw_address"],inplace=True)
```

Creamos una nueva columna que contenga el barrio en el que se localiza cada inmueble.

```python
rstate["new_neigh"] = rstate["neighborhood_id"].str.extract(r'(?<=\:\s)([^(]+)')
```

Concatenamos "raw_address" y new_neigh" en una sola columna, con la que alimentaremos nuestra petición al servidor

```python
rstate["google_address"]  = (rstate['raw_address'] + " " + rstate["new_neigh"] + " " + "Madrid, Spain").replace(" ","+")
```

Definimos la función que utilizaremos para extraer las coordenadas de la API GeoCoding

```python
def get_coordinates(address):
    key = os.getenv('GOOGLEAUTH')
    url = 'https://maps.googleapis.com/maps/api/geocode/json?address='
    try:
        rqst  = requests.get(url + address + '&key=' + key)    
        datos = rqst.json()
        location = datos['results'][0]['geometry']['location']
        print(url + address + '&key=' + key)
        return location
    except:
        return "N/A"
```

Finalmente, extraemos las coordenadas:

```python
lista_google = list(rstate["google_address"])
coord = list()
for l in lista_google:
    coord.append(get_coordinates(l))    
    
```

Guardamos las coordenadas en una columna adicional del dataset

```python
rstate["coordinates"] = coord
```

Eliminamos las columnas que contienen datos no útiles para nuestro análisis o que no cuentan con suficientes datos para ser tenidas en cuenta.

```python
clean_rstate = rstate.drop(columns=["Unnamed: 0","title","subtitle",
                                    "latitude","longitude",
                                    "raw_address","is_exact_address_hidden",
                                    "street_name","street_number",
                                    "portal","door","neighborhood_id",
                                    "rent_price_by_area","is_rent_price_known",
                                    "is_buy_price_known","are_pets_allowed",
                                    "is_furnished","is_kitchen_equipped",
                                    "has_private_parking","has_public_parking",
                                    "is_orientation_north","is_orientation_west",
                                    "is_orientation_east","is_orientation_south",
                                    "n_floors","sq_mt_allotment","operation",
                                    "energy_certificate","sq_mt_useful","rent_price",
                                    "built_year","has_central_heating",
                                    "has_individual_heating", "has_ac",
                                    "has_fitted_wardrobes","has_garden",
                                    "has_pool","has_terrace","has_balcony",
                                    "has_storage_room", "is_accessible",
                                    "has_green_zones", "is_parking_included_in_price",
                                    "google_address","new_neigh","parking_price",
                                    "house_type_id", "is_floor_under"
                                     ])
```

Exportamos la primera versión del dataset enriquecida con los datos geográficos

```python
clean_rstate.to_csv("outputs/raw_data_with_geocoding.csv")
```
