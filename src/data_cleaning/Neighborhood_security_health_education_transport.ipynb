```python
import pandas as pd
```

En este Jupyter generamos 4 categorías que consideramos que suponen un valor a tener en cuenta a la hora de elegir la ubicación de compra de un nuevo inmueble. Estos datos han sido extraidos del portal de datos oficial del Ayuntamiento de Madrid (https://datos.madrid.es/portal/site/egob/). Hemos agrupado los datos recogidos en las siguientes categorías:
- Seguridad:Comisarías de policía local, comisarías de policía nacional y puestos de bomberos.
- Transporte: Estaciones de metro y estaciones de autobus.
- Salud: Farmacias, hospitales y centros de salud.
- Educación: Centros de enseñanza privados y públicos.    

Estas categorías se tendrán en cuenta posteriormente  para alimentar el modelo de predición de precio.

Hemos recogido estos datos en el siguiente csv, que procedemos a tratar para agrupar en las categorías mencionadas:

```python
df = pd.read_csv("a/Inm Estadisticas_csv.csv",delimiter=";")
df.head(1)
```

```python
df.columns
```

```python
df["security"] = df.loc[:,["Pol Local",'PoliciaNac','Bomberos']].sum(axis=1)
df["transport"] = df.loc[:,["Bocas de m",'AutobusesU']].sum(axis=1)
df["health"] = df.loc[:,["Farmacias",'Hospitales','CentrosSal']].sum(axis=1)
df["education"] = df.loc[:,["CentrosEns",'ColegiosPu']].sum(axis=1)
```

```python
df = df.drop(columns=['OBJECTID *', 'Join_Count', 'TARGET_FID', 'Join_Count.1',
       'TARGET_FID.1', 'F1', 'OBJECTID', 'geodb_oid', 'CODDIS', 'NOMDIS', 'CODBAR',
       'NOMBRE', 'ORIG_FID', 'OBJECTID.1','Pol Local', 'Supermerca', 'PoliciaNac', 'Bomberos', 'AutobusesU',
       'Bocas de m', 'Farmacias', 'Hospitales', 'CentrosSal', 'CentrosEns',
       'ColegiosPu','CODDISTRIT', 'CODBARRIO']).set_index("id")
```
