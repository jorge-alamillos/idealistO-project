```python
import pandas as pd
```

```python
df = pd.read_csv("a/Inm Estadisticas_csv.csv",delimiter=";")
df.head(1)
```

```python
df.columns
```

```python
df["security"] = df.loc[:,["Pol Local",'PoliciaNac','Bomberos']].sum(axis=1)
df["transport"] = df.loc[:,["Bocas de m",'AutobusesU']].sum(axis=1)
df["health"] = df.loc[:,["Farmacias",'Hospitales','CentrosSal']].sum(axis=1)
df["education"] = df.loc[:,["CentrosEns",'ColegiosPu']].sum(axis=1)
```

```python
df = df.drop(columns=['OBJECTID *', 'Join_Count', 'TARGET_FID', 'Join_Count.1',
       'TARGET_FID.1', 'F1', 'OBJECTID', 'geodb_oid', 'CODDIS', 'NOMDIS', 'CODBAR',
       'NOMBRE', 'ORIG_FID', 'OBJECTID.1','Pol Local', 'Supermerca', 'PoliciaNac', 'Bomberos', 'AutobusesU',
       'Bocas de m', 'Farmacias', 'Hospitales', 'CentrosSal', 'CentrosEns',
       'ColegiosPu','CODDISTRIT', 'CODBARRIO']).set_index("id")
```
